```python
import pandas as pd
import numpy as np
df=pd.read_csv(r"C:\Users\ASUS\Downloads\archive (8)\ML_transform_alz.csv")
df.columns
```

```python
df.isnull().sum()
df.dropna()
df.drop_duplicates()
df.drop(columns=['Unnamed: 0', 'Data_Value','Class_code','LocationDesc_code','Stratification2_nan','QuestionCode_women_paptest_3yrs','QuestionCode_women_mammogram_2yrs','QuestionCode_at_risk_pneumococcal_vaccine_ever','StratificationCategory1_nan'], errors='ignore', inplace=True)
print(df.columns)
```

```python
df['mental_risk']=df['QuestionCode_lifetime_depression_diagnosis']+df['QuestionCode_freq_mental_distress']
df['physical_risk']=df['QuestionCode_current_obesity_bmi_30plus']+df['QuestionCode_physically_unhealthy_days_mean']+df['QuestionCode_ever_told_high_blood_pressure']+df['QuestionCode_fair_or_poor_health_with_arthritis']+df['QuestionCode_ever_diagnosed_arthritis']
df['Lifestyle']=df['QuestionCode_no_leisure_time_physical_activity']+df['QuestionCode_sufficient_sleep_over_6h']+df['QuestionCode_binge_drinking_past_30d']+df['QuestionCode_current_smoker_ever_100cigs']
df.drop(columns=['QuestionCode_lifetime_depression_diagnosis','QuestionCode_freq_mental_distress','QuestionCode_current_obesity_bmi_30plus','QuestionCode_physically_unhealthy_days_mean','QuestionCode_ever_told_high_blood_pressure','QuestionCode_fair_or_poor_health_with_arthritis','QuestionCode_ever_diagnosed_arthritis','QuestionCode_no_leisure_time_physical_activity','QuestionCode_sufficient_sleep_over_6h','QuestionCode_binge_drinking_past_30d','QuestionCode_current_smoker_ever_100cigs'],inplace=True)
df.head(10)
df.columns
```

```python
cols_to_drop = [
    
    'QuestionCode_at_risk_pneumococcal_vaccine_ever',
    'QuestionCode_flu_vaccine_past_year',
    'QuestionCode_colorectal_screening_upto_date',
    'QuestionCode_men_up_to_date_clinical_prevention',
    'QuestionCode_women_up_to_date_clinical_prevention',


    'QuestionCode_lost_5_or_fewer_teeth',


    'QuestionCode_no_diabetes_test_within_3yrs',
    'QuestionCode_not_currently_care_expect_care_2yrs',


    'QuestionCode_nan',
    'StratificationCategory1_nan',
    'Stratification1_nan',
    'StratificationCategory2_nan',


    'Stratification1_65 years or older',
    'Stratification1_Overall',
    'StratificationCategory2_Sex',
    'Stratification2_Male',
    'Stratification2_Female',
    'Stratification2_Black, non-Hispanic',
    'Stratification2_Hispanic',
    'Stratification2_White, non-Hispanic',
    'Stratification2_Native Am/Alaskan Native'
]


df = df.drop(columns=[c for c in cols_to_drop if c in df.columns])

print("Remaining columns:", len(df.columns))
print(df.columns)
```

```python
df['cognitive_score']=df[
    [
     'QuestionCode_scd_discussed_with_provider',
        'QuestionCode_scd_interferes_social_household',
        'QuestionCode_scd_needs_daily_assistance',
    ] 
].sum(axis=1)

df['disability_score']=df[
    [
        'QuestionCode_reported_disability_sensory_mobility',
        'QuestionCode_severe_joint_pain_with_arthritis',
        'QuestionCode_fall_injury_past_year'
    ]
].sum(axis=1)

df['care_score']=df[
    [
        'QuestionCode_care_20plus_hours_per_week',
        'QuestionCode_provided_care_6months_plus',
        'QuestionCode_provided_care_past_month',
        'QuestionCode_provided_care_dementia_past_month'
    ]
].sum(axis=1)

df["lifestyle_bio_score"] = df[
    [
        'QuestionCode_eat_2plus_fruits_daily',
        'QuestionCode_eat_3plus_vegetables_daily',
        'QuestionCode_cholesterol_screen_5yrs',
        'QuestionCode_on_bp_meds_if_high_bp',
        'mental_risk',
        'physical_risk',
        'Lifestyle'
    ]
].sum(axis=1)

```

```python
features_to_keep=[
    'cognitive_score','disability_score','care_score','lifestyle_bio_score','QuestionCode_scd_worsening_past_year'
]
df=df[features_to_keep]
df.columns
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix,accuracy_score

X=df[['cognitive_score', 'disability_score', 'care_score','lifestyle_bio_score']]
Y=df['QuestionCode_scd_worsening_past_year']
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25,random_state=42)
model=LogisticRegression(max_iter=1000)
model.fit(X_train,Y_train)
Y_pred=model.predict(X_test)
accuracy=accuracy_score(Y_test,Y_pred)
conf_matrix=confusion_matrix(Y_test,Y_pred)
print(accuracy)
print(conf_matrix)
df['QuestionCode_scd_worsening_past_year'].value_counts(normalize=True)
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix,accuracy_score

X=df[['cognitive_score', 'disability_score', 'care_score','lifestyle_bio_score']]
Y=df['QuestionCode_scd_worsening_past_year']
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25,random_state=42)
model=RandomForestClassifier()
param_grid={
    'n_estimators': [100],     
    'max_depth': [None, 5],          
    'min_samples_split': [2, 5]
    }
grid_search=GridSearchCV(
    estimator=model,
    param_grid=param_grid,
    cv=5,
    n_jobs=-1,
    verbose=1
)
grid_search.fit(X_train,Y_train)
print("Best Parameters:", grid_search.best_params_)
best_model = grid_search.best_estimator_
Y_pred=best_model.predict(X_test)
accuracy=accuracy_score(Y_test,Y_pred)
conf_matrix=confusion_matrix(Y_test,Y_pred)
print(accuracy)
print(conf_matrix)
df['QuestionCode_scd_worsening_past_year'].value_counts(normalize=True)
```
